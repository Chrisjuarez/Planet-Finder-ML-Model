```python
from pathlib import Path
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import warnings
warnings.filterwarnings('ignore')

# Set style for better plots
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (12, 6)

# Load processed data
X = pd.read_parquet("table_v1.parquet")
y = pd.read_csv("labels_v1.csv")
spl = pd.read_csv("split_v1.csv")

# Merge all
df = X.merge(y, on="kepid").merge(spl, on="kepid")

FEATS = [c for c in X.columns if c != "kepid"]
TARGET = "label_lenient"

print("="*60)
print("DATA QUALITY CHECKS")
print("="*60)
print(f"Dataset shape: {df.shape}")
print(f"Number of features: {len(FEATS)}")
print(f"Positive rate: {df[TARGET].mean():.4f}")
```

```python
print("\nFEATURE INVENTORY CHECK")
print("="*60)

# Expected original features
original_features = [
    "teff", "logg", "feh", "radius", "mass", "kepmag",
    "rrmscdpp03p0", "rrmscdpp06p0", "rrmscdpp12p0",
    "dataspan", "dutycycle", "nquarters"
]

# Expected new features
expected_new = [
    "feh_x_teff", "radius_x_kepmag", "transit_prob", "stellar_density",
    "noise_star_ratio", "noise_consistency", "obs_quality", "detection_eff",
    "rrmscdpp03p0_log", "rrmscdpp06p0_log", "rrmscdpp12p0_log"
]

# Check what we actually have
found_original = [f for f in original_features if f in FEATS]
found_new = [f for f in expected_new if f in FEATS]
unexpected = [f for f in FEATS if f not in original_features + expected_new]

print(f"Original features found: {len(found_original)}/{len(original_features)}")
print(f"New engineered features found: {len(found_new)}/{len(expected_new)}")
print(f"Total features: {len(FEATS)}")

if len(found_original) < len(original_features):
    missing = set(original_features) - set(found_original)
    print(f" Missing original features: {missing}")

if len(found_new) < len(expected_new):
    missing_new = set(expected_new) - set(found_new)
    print(f" Missing engineered features: {missing_new}")

if unexpected:
    print(f" Additional features found: {unexpected[:5]}...")  # Show first 5

print("\n Feature engineering success!" if len(found_new) > 5 else "âš ï¸ Few engineered features found")
```

```python
def corr_with_label(frame, features, label_col):   #which relate to planet presence
                                                   #positive r = more likely to have planet
    out = []
    y = frame[label_col].astype(float)
    for c in features:
        x = frame[c].astype(float)
        mask = x.notna() & y.notna()
        if mask.sum() < 3:
            out.append((c, np.nan, np.nan, mask.sum()))
            continue
        r_pearson = np.corrcoef(x[mask], y[mask])[0,1]
        rx = x[mask].rank(method="average")
        ry = y[mask].rank(method="average")
        r_spearman = np.corrcoef(rx, ry)[0,1]
        out.append((c, r_pearson, r_spearman, mask.sum()))
    res = pd.DataFrame(out, columns=["feature","r_pearson_with_label","r_spearman_with_label","n"])
    return res.sort_values(by="r_pearson_with_label", ascending=False)

corr_table = corr_with_label(df, num_feature_cols, LABEL_COL)
corr_table.head(20)
```

```python
top_k = 20
top_abs = corr_table.assign(abs_r=lambda d: d["r_pearson_with_label"].abs()) \
                    .nlargest(top_k, "abs_r")

fig = plt.figure(figsize=(6, 6))
ypos = np.arange(len(top_abs))
plt.barh(ypos, top_abs["r_pearson_with_label"])
plt.yticks(ypos, top_abs["feature"])
plt.xlabel("Pearson r with label_lenient")
plt.title(f"Top {top_k} |Featureâ€“Label| Correlations")
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

```python
def plot_detection_rate_by_feature(frame, feature, label_col, bins=20, split=None):
    sub = frame if split is None else frame[frame[SPLIT_COL] == split]
    x = sub[feature]
    y = sub[label_col]
    mask = x.notna() & y.notna()
    x = x[mask]; y = y[mask]         #answers if relationship is linear of nonlinear or mono

    edges = np.linspace(x.min(), x.max(), bins + 1)
    idx = np.clip(np.digitize(x, edges) - 1, 0, bins - 1)

    det_rate = []
    centers = []
    counts = []
    for b in range(bins):
        m = idx == b
        det_rate.append(y[m].mean() if m.sum() else np.nan)
        centers.append((edges[b] + edges[b+1]) / 2.0)
        counts.append(m.sum())

    fig = plt.figure(figsize=(6, 4))
    plt.plot(centers, det_rate, marker='o')
    plt.xlabel(feature)
    plt.ylabel("Detection rate (mean of label_lenient)")
    title_split = "" if split is None else f" [{split}]"
    plt.title(f"Detection Rate vs {feature}{title_split}")
    plt.grid(True)
    plt.tight_layout()
    plt.show()

    return pd.DataFrame({"center": centers, "detection_rate": det_rate, "count": counts})

for feat in ["duty_cycle", "dataspan", "teff", "logg", "feh"]:
    if feat in df.columns:
        _ = plot_detection_rate_by_feature(df, feat, LABEL_COL, bins=25, split="train")
```

```python
def detection_rate_2d(frame, f1, f2, label_col, bins=20, split=None):
    sub = frame if split is None else frame[frame[SPLIT_COL] == split]
    x = sub[f1]; y = sub[f2]; z = sub[label_col]
    mask = x.notna() & y.notna() & z.notna()
    x = x[mask].to_numpy()
    y = y[mask].to_numpy()
    z = z[mask].to_numpy() #do 2 feats interact interestingly? look for bright

    x_edges = np.linspace(x.min(), x.max(), bins + 1)
    y_edges = np.linspace(y.min(), y.max(), bins + 1)

    xi = np.clip(np.digitize(x, x_edges) - 1, 0, bins - 1)
    yi = np.clip(np.digitize(y, y_edges) - 1, 0, bins - 1)

    sums = np.zeros((bins, bins), dtype=float)
    cnts = np.zeros((bins, bins), dtype=int)
    for a, b, label in zip(xi, yi, z):
        sums[b, a] += label
        cnts[b, a] += 1
    with np.errstate(invalid='ignore', divide='ignore'):
        rate = sums / cnts
    rate[cnts == 0] = np.nan

    fig = plt.figure(figsize=(6, 5))
    plt.imshow(rate, origin='lower', aspect='auto',
               extent=[x_edges[0], x_edges[-1], y_edges[0], y_edges[-1]],
               interpolation='nearest')
    cbar = plt.colorbar()
    cbar.set_label("Detection rate")
    plt.xlabel(f1)
    plt.ylabel(f2)
    title_split = "" if split is None else f" [{split}]"
    plt.title(f"2D Detection Rate: {f1} Ã— {f2}{title_split}")
    plt.tight_layout()
    plt.show()

    return rate, x_edges, y_edges, cnts
pairs = [
    ("duty_cycle", "dataspan"),
    ("teff", "logg"),
    ("feh", "teff"),
]
for f1, f2 in pairs:
    if f1 in df.columns and f2 in df.columns:
        _ = detection_rate_2d(df, f1, f2, LABEL_COL, bins=24, split="train")
#planets seem to go for cooler stars with more metals
```
